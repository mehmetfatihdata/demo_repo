## 2: Plot 12 most dangerous cities

```python
kill1 = kill.dropna()
```

```python
dang_city = pd.DataFrame(kill1['city'].value_counts()[:12])
dang_city
```

```python
plt.figure(figsize= (8,5))
sns.barplot(data = dang_city, x = dang_city.index, y = 'city')
plt.xticks(rotation = 45)
plt.ylabel('')
plt.title('Most dangerous cities', fontsize = 15, color = 'b')
plt.show()
```

```python
# with seaborn barplot

```

```python
plt.figure(figsize= (8,5))
sns.countplot(data = kill1, x = 'city', order = kill1['city'].value_counts()[:12].index)
plt.xticks(rotation = 45)
plt.ylabel('')
plt.title('Most dangerous cities', fontsize = 15, color = 'b')
plt.show()
```

```python
# with seaborn countplot
```

## 3: Plot 20 most dangerous states
