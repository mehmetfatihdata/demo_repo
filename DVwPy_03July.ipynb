```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
print(sns.get_dataset_names())
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.countplot(x ='day', data = tips);
```

```python
#  seabornda matplotlib uygulaması yapacak isek ax parametresini kullanmalıyız.

fig, ax = plt.subplots()

# veya sns.countplot(x = 'day', data = tips, ax = ax)
ax = sns.countplot(x = 'day', data = tips);

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+1));
```

```python
sns.countplot(tips['day']);
```

```python
fig, ax_tip = plt.subplots(figsize = (10,6))
ax_tip = sns.countplot(x="day", data = tips, hue = 'smoker');
# sns.countplot(x="day", data = tips, ax =ax_tip );
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+1));
    
```

```python
fig, ax_tip = plt.subplots(figsize = (10,6))
ax_tip = sns.countplot(x="day", data = tips, hue = 'size');
# sns.countplot(x="day", data = tips, ax =ax_tip );
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.05, p.get_height()+1));
    
```

```python
fig, ax_tip = plt.subplots()
ax_tip = sns.countplot(x="day", data = tips);
# sns.countplot(x="day", data = tips, ax = ax_tip );
for p in ax_tip.patches:
    
    ax_tip.annotate((p.get_height()), (p.get_x()+0.3, p.get_height()+1))
    ax_tip.annotate('%' + str(round((p.get_height())/tips.day.count(), 2)*100), (p.get_x()+0.15, p.get_height()-10));
    
```

```python
fig, ax_tip = plt.subplots()
ax_tip = sns.countplot(x="day", data = tips);
# sns.countplot(x="day", data = tips, ax = ax_tip );
for p in ax_tip.patches:
    ax_tip.annotate('(' +
                    str(p.get_height()) +
                    ') ' +
                     '%' +
                    str(round((p.get_height())/tips.day.count(), 2)*100), (p.get_x()+0.05, p.get_height()-10));
    
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
plt.figure(figsize = (12,6))
sns.lineplot(data = flights, x = "year", y = "passengers")
plt.show()
```

```python
flights.groupby(['year', 'month']).sum()
```

```python
plt.figure(figsize = (12, 6))
sns.lineplot(y = flights.passengers, x = flights.index)
plt.show()
```

```python
flights_wide = flights.pivot("year", "month", "passengers")
flights_wide.head()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=flights_wide.T);
```

```python
sns.lineplot(data = flights, x = 'month', y = 'passengers', hue='year')
plt.legend(loc=(1.1,.0));
```

```python
plt.figure(figsize = (12,6))
sns.pointplot(data = flights, x = 'month', y = 'passengers', hue='year', palette = "viridis")  # hue ekledik, palette ekledik
plt.legend(loc=(1.1,.0))
plt.show()
```
