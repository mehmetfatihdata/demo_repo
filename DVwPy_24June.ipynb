```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y, color ="darkred", linewidth = 5, ls = '--')
ax.plot(x,y-1, color ="darkblue", lw = 6, ls = '-.')
ax.plot(x,y-2, color ="yellow", lw = 3, ls = ':')
ax.plot(x,y-3, color ="black", lw =2, ls = '-')
```

```python
fig, ax = plt.subplots(figsize=(6,4))
ax.plot(x,y, marker="+", markersize=10)
ax.plot(x,y-1, marker="o", markersize=10)
ax.plot(x,y-2, marker="s", markersize=10)
ax.plot(x,y-3, marker="1", markersize=10)
```

```python
x = np.arange(1,10)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
ax[0].plot(x,x**2,x,x**3,x,x**4)
ax[1].plot(x,x**2,x,x**3,x,x**4)
ax[1].set_xlim([2,4])
ax[1].set_ylim([9,100])

ax[0].axvline(x = 3, ls = '--')
ax[0].axhline(y = 50, ls = '--')

ax[1].axvline(x = 3, ls = '--')
ax[1].axhline(y = 50, ls = '--')
```
