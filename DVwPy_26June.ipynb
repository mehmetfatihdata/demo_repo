```python
def multi_graf(v1 = age, v2 = salary, rows = 2, cols = 3):
    fig, ax = plt.subplots(nrows=rows, ncols=cols)
    for i in range(rows):
        for j in range(cols):    
            ax[i,j].plot(v1,v2)
            plt.tight_layout()
            
multi_graf()
```

```python
day = tips.day.unique()
day
```

```python
total_bill = tips.groupby('day')['total_bill'].sum()
total_bill
```

```python
days = total_bill.index
```

```python
fig, ax = plt.subplots()

ax.bar(days, total_bill)
plt.show()
```

```python
day = tips.groupby("sex")["total_bill"].sum().index
tip = tips.groupby("sex")["tip"].sum()
fig, ax = plt.subplots()
ax.bar(day,tip)
plt.show()
```

```python
day = tips.groupby("sex")["total_bill"].sum().index
tip = tips.groupby("sex")["tip"].sum()
total = tips.groupby()
fig, ax = plt.subplots()
ax.bar(day,tip, label ='tip')
ax.bar(day,total, label = 'total')
plt.show()
```

```python
day = tips.groupby("day")["total_bill"].sum().index
tip = tips.groupby("day")["tip"].sum()
total = tips.groupby("day")["total_bill"].sum()
fig, ax = plt.subplots()

ax.bar(day,tip, label="tip")
ax.bar(day,total, label="total", width = 0.4, align ='edge')

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```
