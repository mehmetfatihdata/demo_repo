```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Set darkgrid style from seaborn

Bar plots

```python
income = [100, 80, 150, 48, 52, 69, 88]
expense = [30, 100, 100, 20, 75, 50, 28]
index = ['A', 'B', 'C','D', 'E', 'F', 'G']
df = pd.DataFrame({'income': income,'expense': expense}, index=index)
df.head()
```

# 2. Make a bar plot

```python
df = pd.melt(df, id_vars="class", var_name="sex", value_name="survival rate")
df
```

```python
sns.barplot(data = df, x = df.index, y = 'income', color = 'blue')
sns.barplot(data = df, x = df.index, y = 'expense', color = 'orange')
plt.legend(labels = ['income', 'expense'])
```

# 6. Histogram of the total_bill column

```python
sns.set_theme(style="darkgrid")
sns.histplot(data = tips, x = 'total_bill', color =  '#0066ff', bins = 10)
plt.xlabel('')
plt.ylabel('Frequency')
plt.show()
```

# 10. Set an additional dimension using size column

```python
ax = sns.scatterplot(data = tips, x = 'tip', y = 'total_bill', hue = 'size', palette = 'coolwarm')
norm = plt.Normalize(tips['size'].min(), tips['size'].max())
sm = plt.cm.ScalarMappable(cmap="coolwarm", norm=norm)
sm.set_array([])
ax.get_legend().remove()
ax.figure.colorbar(sm)
plt.xlabel('')
plt.show()
```
